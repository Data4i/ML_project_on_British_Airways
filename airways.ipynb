```python
df['is_verified'].value_counts().plot(kind = 'bar')
```

```python
def sort_is_verified(row):
    if row == 'Verified Review' or row == 'Trip Verified':
        return 'Verified'
    elif row == 'Not Verified' or row == 'Unverified':
        return 'Unverified'
    elif row == 'not specified':
        return 'Not Specified'
    else:
        return 'other'
    
df['is_verified'] = df['is_verified'].apply(sort_is_verified)
```

```python
df = df[df['is_verified'] != 'other']
```

```python
df['is_verified'].value_counts().plot(kind = 'bar')
```

```python
df['no_of_reviews'].value_counts().plot(kind = 'bar')
```

```python
def restructure_no_of_reviews(row):
    if row == 0:
        return 'one'
    return 'more than one'

# df['is_verified'] = 
df['no_of_reviews'] = df['no_of_reviews'].apply(restructure_no_of_reviews)
```
